```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re # regular expression useful for punctution or to perform repetetive task
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
    # lowercase
    text =str(text).lower()
    # removed the punctutaion
    text = re.sub(r'[^a-zA-Z\s]','',text)
    words = text.split() # word_tokenization
    # remove stopwords
    words = [word for word in words if word not in stop_words]
    # lemmatize
    words = [lemmatizer.lemmatize(word) for word in words]

    return " ".join(words)
```

```python
df['tags']= df['tags'].apply(preprocess_text)
```

```python
df['tags'][1]
```

```python
df = df.reset_index(drop = True)
```

```python
indices = pd.Series(df.index,index = df['title']).drop_duplicates()
```

```python
indices
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features=50000,ngram_range=(1,2),stop_words='english')
```

```python
tfidf_matrix = tfidf.fit_transform(df['tags'])
```

```python
tfidf_matrix 
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend(title, n = 10):
  if title not in indices:
    return ['Movie not found']

  idx = indices[title]
  sim_score = cosine_similarity(tfidf_matrix[idx], tfidf_matrix).flatten()
  similar_idx = sim_score.argsort()[::-1][1:n+1]
  return df['title'].iloc[similar_idx]
```

```python
recommend('Avatar',5)
```

```python
recommend('Toy Story')
```
